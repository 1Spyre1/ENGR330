```python
def snap_to_025(x):
    return (np.round(x / 0.025) * 0.025).round(3)
```

```python
def build_datetime(row):
    if pd.isna(row.DISCOVERY_TIME):
        hour = 12
        minute = 0
    else:
        t = int(row.DISCOVERY_TIME)
        hour = t // 100
        minute = t % 100

    date = pd.to_datetime(row.DISCOVERY_DATE)
    return date.replace(hour=hour, minute=minute)
```

```python
try1["DISCOVERY_DATE"] = pd.to_datetime(try1["DISCOVERY_DATE"], utc=True)
```

```python
time = try1["DISCOVERY_TIME"].fillna(1200).astype(int)

hours = time // 100
minutes = time % 100
```

```python
try1["datetime"] = (
    try1["DISCOVERY_DATE"]
    + pd.to_timedelta(hours, unit="h")
    + pd.to_timedelta(minutes, unit="m")
)
try1["datetime"]
```
